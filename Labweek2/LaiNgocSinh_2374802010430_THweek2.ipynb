```python
def get_mean(score):
  return sum(score)/len(score)
```

```python
def Q1(g):
  return np.percentile(g, 25)
```

```python
def Q2(g):
  return np.percentile(g, 50)
```

```python
def Q3(g):
  return np.percentile(g, 75)
```

```python
def std(g):
  return np.std(g)
```

```python
pd.pivot_table(data=dftuyensinh,index='KT', values='DH1',
               aggfunc=['sum', 'mean','median','min','max', std, Q1, Q2, Q3])
```
